```python
#kadai_005

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

#　折れ線グラフ
x0 =[0,1,2,3,4]
y0 =[1, -5, 2, 1, 2]
# プロット
axes[0,0].plot(y0)

#　sin関数グラフ
x1 = np.linspace(0,10,100)
y1 = 2+np.sin(2*x1)
# プロット
axes[0,1].plot(x1, y1, color="red", linewidth=2)


#　ヒストグラム
x2 = np.random.normal(15,5,2000)
# プロット
axes[0,2].hist(x2,color="green", bins=30)

# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0,2,60)
y3 = 4 + np.random.normal(0,2,len(x3))
# プロット
axes[1,0].scatter(x3,y3)


# 箱ひげ図
# モーターの周波数
hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]
axes[1,1].boxplot(hertz)



plt.show()
```
